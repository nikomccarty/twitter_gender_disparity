friends_count, in this dataset, is actually the number of people that they are FOLLOWING.

followers_count is exactly what it sounds like -- the number of followers that each person has.

```python
df.rename(columns={"friends_count": "following_count"}, inplace=True)
```

```python
df.head()
```
